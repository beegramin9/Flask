```python
park = pd.read_csv('./static/data/park/공원 위치, 면적, 인구.csv',
                       index_col='공원', encoding='utf8')
park.head()
```

```python
import folium
```

```python
park_name = '남산도시자연공원'
```

```python
selected = park[park.index == park_name]
rest = park[park.index != park_name]

selected.index[0]
```

```python
selected['지역'][0]
```

```python
selected['공원개요'][0]
```

```python
selected['면적'][0]
```

```python
mapping = folium.Map([37.5502, 126.982], zoom_start=11)

for i in selected.index:
    folium.CircleMarker( [selected.lat[i], selected.lng[i]] , radius = float(selected['면적 코드'][i]) * 5, color = 'green', fill_color = 'green').add_to(mapping)

mapping
```
